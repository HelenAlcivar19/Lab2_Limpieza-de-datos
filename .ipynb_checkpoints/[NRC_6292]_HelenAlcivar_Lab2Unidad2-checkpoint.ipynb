### Error 1 ElinaciÃ³n de la colunma Range

```python
'''Ocupamos el comando drop para eliminar la colunma Range'''
data = df.drop(['Range'], axis=1)
data
```

Tal y como se muestra el segundo error trata de una colunma con valores NaN este es un valor innecesario que debe eliminarse para analizar correctamente el conjunto de datos.Se detecto el problema por que al momento de cargar los datos del dataset se visualiza la colunma Range posee puros datos Nan,entonces se decidio eliminar por que no influye en nuestro dataset.

### Error 2 Exceso de mayusculas y minusculas

```python
'''Realizamos el exceso de las mayusculas y minusculas en la colunma de City, donde existen
registros con nombres: San Francisco,S,0 y seran sustituidos por SAN FRANCISCO con el comando  replace '''
data['City'] = data['City'].replace('San Francisco','SAN FRANCISCO')
data['City'] = data['City'].replace('S','SAN FRANCISCO')
data['City'] = data['City'].replace('NA','SAN FRANCISCO')
data.groupby('City').count()
```

Tal y como se muestra el segundo error trata de el exceso de mayuscula y minuscula en el dataset.Se detecto el problema por que al momento de cargar los datos del dataset se lo visualiza en la colunma city registros con nombres como : San Francisco,S,0 y seran sustituidos por el nombre de SAN FRANCISCO.El principal problema de estos registros mal estructurados es que al momento de hacer una consulta el programa no se encuentre capaz de reconocer la informacion y de errores. Por ende es mejor trabajar con un mismo tipo de letra y asi no tener complicaciones al momento de hacer alguna consulta.
